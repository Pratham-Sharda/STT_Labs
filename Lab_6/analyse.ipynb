```python
# Re-import necessary libraries after execution reset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define sequential execution time
sequential_time = 5.06

# Define test configurations with their average execution times and failures
test_data = {
    "pytest_n_auto_dist_load": (10.07, 0.00),
    "pytest_n_auto_dist_load_parallel_1": (19.38, 0.00),
    "pytest_n_auto_dist_load_parallel_auto": (61.49, 4.00),
    "pytest_n_auto_dist_no": (10.78, 0.00),
    "pytest_n_auto_dist_no_parallel_1": (13.76, 0.00),
    "pytest_n_auto_dist_no_parallel_auto": (51.50, 4.00),
    "pytest_n1_dist_load": (6.73, 0.00),
    "pytest_n1_dist_load_parallel_1": (7.14, 0.00),
    "pytest_n1_dist_load_parallel_auto": (53.06, 4.00),
    "pytest_n1_dist_no": (8.29, 0.00),
    "pytest_n1_dist_no_parallel_1": (11.65, 0.00),
    "pytest_n1_dist_no_parallel_auto": (52.87, 4.00),
    "pytest_parallel_threads_1": (7.18, 0.00),
    "pytest_parallel_threads_auto": (35.03, 4.00)
}
# Compute speedup ratios
execution_matrix = []
for config, (time, failures) in test_data.items():
    speedup = sequential_time / time
    execution_matrix.append([config, time, failures, round(speedup, 3)])
execution_df = pd.DataFrame(execution_matrix, columns=["Test Configuration", "Avg Execution Time (s)", "Failures", "Speedup Ratio"])
execution_df = execution_df.sort_values(by="Speedup Ratio", ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=execution_df["Speedup Ratio"], y=execution_df["Test Configuration"], palette="Blues_r")
plt.xlabel("Speedup Ratio")
plt.ylabel("Test Configuration")
plt.title("Speedup Ratios for Different Parallelization Modes")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
# Display the execution matrix
execution_df
```
